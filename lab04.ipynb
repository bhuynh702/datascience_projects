```python
#Figure 2
steps.select('Index','Minutes Very Active','Minutes Fairly Active','Minutes Lightly Active','Minutes Sedentary').plot('Index')
# The pattern I notice is that on most days Minutes Sedentary is the highest, and Minutes Lightly Active is the second highest.
```
